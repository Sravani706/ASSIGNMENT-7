```python
# -*- coding: utf-8 -*-
"""Untitled16.ipynb
Automatically generated by Colaboratory.
Original file is located at
    https://colab.research.google.com/drive/1ER5TC9ISyC2j5AuAKGXfGLXbdlYG2ZhU
"""

import numpy as np
import matplotlib.pyplot as plt

#Generate line points
def line_gen(A,B):
  len =10
  dim = A.shape[0]
  x_AB = np.zeros((dim,len))
  lam_1 = np.linspace(0,1,len)
  for i in range(len):
    temp1 = A + lam_1[i]*(B-A)
    x_AB[:,i]= temp1.T
  return x_AB

#Generating and plotting line x+0.5y=6
A=np.array([20,0])
B=np.array([0,15])
AB=line_gen(A,B)
plt.plot(AB[0,:],AB[1,:],label='x+0.5y=6')

#Generating and plotting line 0.3x+2.25y=9
C=np.array([0,10])
D=np.array([30,0])
CD=line_gen(C,D)
plt.plot(CD[0,:],CD[1,:],label='0.3x+2.25y=9')

#Generating and plotting line 2x+3y=24
E=np.array([10,0])
F=np.array([0,30])
EF=line_gen(E,F)
plt.plot(EF[0,:],EF[1,:],label='2x+3y=24')


plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend(loc='best')
plt.grid() # minor
plt.axis('equal')
plt.show()
```
